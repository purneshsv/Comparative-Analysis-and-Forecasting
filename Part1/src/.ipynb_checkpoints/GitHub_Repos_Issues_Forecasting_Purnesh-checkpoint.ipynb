## REQUIREMENT 1

Use Python/GitHub API to retrieve Issues/Repos information of the **past 2 months** for the following  repositories:
- https://github.com/langchain-ai/langchain
- https://github.com/langchain-ai/langgraph
- https://github.com/microsoft/autogen
- https://github.com/openai/openai-cookbook 
- https://github.com/elastic/elasticsearch 
- https://github.com/milvus-io/pymilvus 


## Important: Data Retrieval Timeframe

 - For **chart visualizations below**, retrieve **past 2 years** of data.
 - For **semantic search and vector embeddings**, use **past 2 months** of data from the above repositories.

    
</div>



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import requests
from datetime import datetime, timedelta

# Your GitHub token for authentication
token = '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'


def get_repo_info(repo_url):
    owner, repo = repo_url.split('/')[-2:]
    issues_url = f"https://api.github.com/repos/{owner}/{repo}/issues"
    two_months_ago = datetime.now() - timedelta(days=730)
    since_date = two_months_ago.strftime("%Y-%m-%dT%H:%M:%SZ")
    params = {'state': 'all', 'since': since_date, 'per_page': 100}
    headers = {
        'Authorization': f'token {token}',
        'Accept': 'application/vnd.github.v3+json'
    }
    all_issues = []
    while True:
        response = requests.get(issues_url, headers=headers, params=params)
        if response.status_code == 200:
            current_page_issues = response.json()
            for issue in current_page_issues:
                issue_date = datetime.strptime(issue['created_at'], "%Y-%m-%dT%H:%M:%SZ")
                if issue_date >= two_months_ago:
                    all_issues.append(issue)
            if 'next' in response.links:
                issues_url = response.links['next']['url']
            else:
                break
        else:
            print(f"Failed to retrieve data for {repo_url}. Status code: {response.status_code}, Response: {response.text}")
            break
    return all_issues

# List of repository URLs
repo_urls = [
    "https://github.com/langchain-ai/langchain",
    "https://github.com/langchain-ai/langgraph",
    "https://github.com/microsoft/autogen",
    "https://github.com/openai/openai-cookbook",
    "https://github.com/elastic/elasticsearch",
    "https://github.com/milvus-io/pymilvus",
]

# Retrieve information for each repository
repo_data = {}
for url in repo_urls:
    repo_data[url] = get_repo_info(url)

# Print the retrieved data for verification
for url, data in repo_data.items():
    print(f"Repository: {url}")
    if data:
        print(f"Issues count: {len(data)}")
    else:
        print("Data retrieval failed.")

```

## REQUIREMENT 2
- A line chart to plot the issues for every repository
    
</div>


```python
def plot_issues_line_chart(repo_data):
    plt.figure(figsize=(10, 5))
    for repo, data in repo_data.items():
        if isinstance(data, list):  # Filter out stats data
            issue_dates = [datetime.strptime(issue['created_at'], "%Y-%m-%dT%H:%M:%SZ") for issue in data]
            plt.plot(sorted(issue_dates), range(1, len(issue_dates) + 1), label=repo.split('/')[-1])

    plt.xlabel("Date")
    plt.ylabel("Cumulative Issues")
    plt.title("Issues Over Time by Repository")
    plt.legend()
    plt.show()

plot_issues_line_chart(repo_data)
```

## REQUIREMENT 3
- A bar chart to plot the issues created for every month for every repository
    
</div>

```python

from collections import defaultdict


def plot_monthly_issues(repo_data):
    monthly_counts = defaultdict(lambda: defaultdict(int))
    for repo, issues in repo_data.items():
        for issue in issues:
            month = datetime.strptime(issue['created_at'], "%Y-%m-%dT%H:%M:%SZ").strftime('%Y-%m')
            monthly_counts[repo][month] += 1

    # Plotting the data
    fig, ax = plt.subplots(figsize=(14, 7))
    months = sorted(set(sum([list(data.keys()) for data in monthly_counts.values()], [])))
    width = 1 / (len(repo_data) + 1)
    x = np.arange(len(months))
    
    for i, (repo, counts) in enumerate(monthly_counts.items()):
        issue_counts = [counts.get(month, 0) for month in months]
        ax.bar(x + i * width, issue_counts, width, label=repo.split('/')[-1])

    ax.set_xlabel('Month')
    ax.set_ylabel('Issues Created')
    ax.set_title('Monthly Issues Created by Repository')
    ax.set_xticks(x + width*(len(repo_data)-1)/2)
    ax.set_xticklabels(months, rotation=45)
    ax.legend()
    plt.show()

plot_monthly_issues(repo_data)
```

## REQUIREMENT 4
- A bar chart to plot the starts for every repository
    
</div>

```python
def plot_repo_stars(repo_urls):
    stars = []
    labels = []
    for url in repo_urls:
        owner, repo = url.split('/')[-2:]
        repo_url = f"https://api.github.com/repos/{owner}/{repo}"
        response = requests.get(repo_url, headers={'Authorization': f'token {token}'})
        if response.status_code == 200:
            repo_data = response.json()
            stars.append(repo_data['stargazers_count'])
            labels.append(repo)
        else:
            print(f"Failed to retrieve stars for {repo}. Status code: {response.status_code}")

    plt.figure(figsize=(10, 5))
    plt.bar(labels, stars, color='blue')
    plt.xlabel('Repository')
    plt.ylabel('Stars')
    plt.title('Number of Stars per Repository')
    plt.xticks(rotation=45)
    plt.show()

plot_repo_stars(repo_urls)
```

## REQUIREMENT 5
- A bar chart to plot the forks for every repository
    
</div>

```python
def plot_repo_forks(repo_urls):
    forks = []
    labels = []
    for url in repo_urls:
        owner, repo = url.split('/')[-2:]
        repo_url = f"https://api.github.com/repos/{owner}/{repo}"
        response = requests.get(repo_url, headers={'Authorization': f'token {token}'})
        if response.status_code == 200:
            repo_data = response.json()
            forks.append(repo_data['forks_count'])
            labels.append(repo)
        else:
            print(f"Failed to retrieve forks for {repo}. Status code: {response.status_code}")

    plt.figure(figsize=(10, 5))
    plt.bar(labels, forks, color='green')
    plt.xlabel('Repository')
    plt.ylabel('Forks')
    plt.title('Number of Forks per Repository')
    plt.xticks(rotation=45)
    plt.show()

plot_repo_forks(repo_urls)
```

## REQUIREMENT 6
- A bar chart to plot the issues closed for every week for every repository
    
</div>

```python
def plot_weekly_closed_issues(repo_data):
    weekly_closed_counts = defaultdict(lambda: defaultdict(int))
    for repo, issues in repo_data.items():
        for issue in issues:
            if issue['state'] == 'closed':
                closed_week = datetime.strptime(issue['closed_at'], "%Y-%m-%dT%H:%M:%SZ").strftime('%Y-W%U')
                weekly_closed_counts[repo][closed_week] += 1

    # Plotting the data
    fig, ax = plt.subplots(figsize=(14, 7))
    weeks = sorted(set(sum([list(data.keys()) for data in weekly_closed_counts.values()], [])))
    width = 1 / (len(repo_data) + 1)
    x = np.arange(len(weeks))
    
    for i, (repo, counts) in enumerate(weekly_closed_counts.items()):
        closed_counts = [counts.get(week, 0) for week in weeks]
        ax.bar(x + i * width, closed_counts, width, label=repo.split('/')[-1])

    ax.set_xlabel('Week')
    ax.set_ylabel('Issues Closed')
    ax.set_title('Weekly Closed Issues by Repository')
    ax.set_xticks(x + width*(len(repo_data)-1)/2)
    ax.set_xticklabels(weeks, rotation=45)
    ax.legend()
    plt.show()

plot_weekly_closed_issues(repo_data)
```

## REQUIREMENT 7
- A stack-bar Chart to plot the created and closed issues for every repository
    
</div>

```python
# #Add your code for requirement 7 in this cell

import matplotlib.pyplot as plt
from collections import defaultdict
import requests
from datetime import datetime, timedelta

# Function to plot a stacked bar chart for created and closed issues
def plot_stack_bar_created_closed_issues(repo_data):
    created_counts = defaultdict(int)
    closed_counts = defaultdict(int)
    for repo, issues in repo_data.items():
        for issue in issues:
            created_counts[repo] += 1  # Count each issue as created
            if issue['state'] == 'closed':
                closed_counts[repo] += 1  # Count closed issues

    labels = [url.split('/')[-1] for url in created_counts.keys()]  # Extracting repo names from URLs
    created_vals = [created_counts[repo] for repo in created_counts.keys()]
    closed_vals = [closed_counts[repo] for repo in created_counts.keys()]

    plt.figure(figsize=(10, 5))
    plt.bar(labels, created_vals, label='Created', color='b')
    plt.bar(labels, closed_vals, bottom=created_vals, label='Closed', color='r')
    plt.xlabel('Repository')
    plt.ylabel('Number of Issues')
    plt.title('Stacked Bar Chart of Created and Closed Issues')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()  # Adjust layout to make room for label rotation
    plt.show()

# Assuming repo_data is already filled with issues data from your repositories
plot_stack_bar_created_closed_issues(repo_data)


                  
```
